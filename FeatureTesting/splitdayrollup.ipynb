```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Import datasets
df_mean = pd.read_csv(r'F:\Sam Butterfield\Desktop\NClSummerInternship\aggregation_datasets\splitday_rollup_mean.csv')

# Remove eid
df_mean = df_mean.iloc[:,1:]
```

```python
mean_corr = df_mean.corr(method = 'pearson')
```

```python
mask = np.triu(np.ones_like(mean_corr, dtype=np.bool))
f, ax = plt.subplots(figsize=(11, 9))
sns.heatmap(mean_corr, mask = mask)
```

```python
columns = np.full((mean_corr.shape[0],), True, dtype=bool)
for i in range(mean_corr.shape[0]):
    for j in range(i+1, mean_corr.shape[0]):
        if (mean_corr.iloc[i,j] >= 0.9):
            if columns[j]:
                columns[j] = False
selected_columns = df_mean.columns[columns]
data = df_mean[selected_columns]
print(len(df_mean.columns))
print(len(df_mean.columns) - len(data.columns))
```
