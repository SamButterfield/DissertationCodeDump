```python
import pandas as pd
import numpy as np

# ===== Preprocessing =====
# Import the HLAF Dataset
df = pd.read_csv(r'F:\Sam Butterfield\Desktop\NClSummerInternship\HLAF_dataframe_84k.csv')
# Chance id to eid
df.rename(columns = {'id' : 'eid'}, inplace = True)
```

```python
# Duplicate Checking
print("HLAF Full")
print(len(df))
hlaf_df = df.drop_duplicates(subset = ['eid'], keep = "first").copy()
print("HLAF Drop duplicates")
print(len(hlaf_df))

```

```python
# ===== Transpose HLAF =====
hlaf_cols = hlaf_df.iloc[:,1:]
eid_cols = hlaf_df.iloc[:,0].to_frame()

eid = []
time = []
activity = []
feature = []
value = []

eid_counter = 0


for index, row in hlaf_cols.iterrows():
    
    for i in hlaf_cols.columns:
        # Add eid
        eid.append(eid_cols.iloc[eid_counter,0])
        # Add Time
        if "morning" in str(i):
            time.append("morning")
        elif "afternoon" in str(i):
            time.append("afternoon")
        elif "evening" in str(i):
            time.append("evening")
        elif "overnight" in str(i):
            time.append("overnight")
        # Add activity
        if "sed" in str(i):
            activity.append("sed")
        elif "light task" in str(i):
            activity.append("light task")
        elif "mod" in str(i):
            activity.append("mod")
        elif "walking" in str(i):
            activity.append("walking")            
        elif "sleep" in str(i):
            activity.append("sleep")
        # Add feature
        if "percent time" in str(i):
            feature.append("percent time")
        elif "number bout" in str(i):
            feature.append("number bout")
        elif "bout length" in str(i):
            feature.append("bout length")
        # Add value
        value.append(hlaf_cols.at[index,i])
    eid_counter = eid_counter + 1
    
            
hlaf_transposed = pd.concat([pd.Series(eid), pd.Series(time), pd.Series(activity), pd.Series(feature), pd.Series(value)], axis = 1, sort = False)
hlaf_transposed.columns = ['eid', 'time', 'activity', 'feature', 'value']
print(len(hlaf_transposed)/60)          

#hlaf_transposed.to_csv(r'F:\Sam Butterfield\Desktop\NClSummerInternship\hlaf_transposed_unique_eid.csv' , index = False)
#print("written to file")
```
